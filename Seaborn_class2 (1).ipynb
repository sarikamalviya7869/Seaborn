# Seaborn – Class 2: Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips
```

```python
sns.barplot(x=tips['day'],y=tips['total_bill'])
```

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'])
```

```python
sns.countplot(x=tips['sex'])
#Mostly used Machine learning for class Imbalance
```

## boxplot()

```python
sns.boxplot(x=tips['sex'], y=tips['total_bill']) #important plot
```

## stripplot() & swarmplot()

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'])
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |


```python
sns.swarmplot(x=tips['day'], y=tips['total_bill'])
```

```python
# Question:-1 How Many smokers V/s Non Smokers Group by Gender

sns.countplot(x=tips['smoker'],hue=tips['sex'])

```

```python
# Question:-2 Average total bill according to smokers
sns.barplot(x=tips['smoker'], y=tips['total_bill'])
```

```python
# QUestion:-3 Average tip by day, Group by time

sns.barplot(x=tips['day'], y=tips['tip'], hue=tips['time'])
```

```python
# Question:-4 Spread of total bill across days

sns.boxplot(x=tips['day'], y=tips['total_bill'])
```

```python
# Question:-5 Tip distribution by smoker, group by gender
sns.boxplot(x=tips['smoker'], y=tips['tip'] ,hue=tips['sex'])
```
